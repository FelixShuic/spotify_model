Датасет разбит на жанры, содержит такие значимые параметры как имя артиста, название альбома, параметр популярности, длительности, наличие нецензурной лексики, музыкальные параметры выраженные численно, количество долей и жанр песни.

track_id – Уникальный идентификатор трека в Spotify. Это строка, которая однозначно идентифицирует трек в базе данных Spotify.

artists – Исполнители трека(строка).

album_name – Название альбома, в котором был выпущен трек(строка).

track_name – Название самого трека(строка).

popularity – Популярность трека на Spotify, оцененная от 0 до 100. Чем выше значение, тем популярнее трек среди пользователей платформы(число).

duration_ms – Длительность трека в миллисекундах(число).

explicit – Флаг, указывающий на наличие нецензурной лексики в треке. Значение True означает, что трек содержит ненормативную лексику, а False — нет.(bool)

danceability – Параметр, описывающий, насколько трек удобен для танцев (от 0 до 1). Чем выше значение, тем легче танцевать под этот трек.(число, float)

energy – Энергичность трека (от 0 до 1). Чем выше значение, тем более динамичный и интенсивный трек.(число, float)

key – Ключ (тональность) трека, выраженный числом от 0 до 11. Это относится к музыкальной шкале и указывает на главную тональность трека (например, 0 – это C, 1 – C#/Db и т.д.).(число)

loudness – Общая громкость трека в децибелах (dB). Это значение отражает громкость, при которой трек был записан и воспроизводится.(отрицательное число, float)

mode – Модальность трека, которая определяет его тональность как мажорную (1) или минорную (0). (int bool)

speechiness – Параметр, указывающий, насколько трек содержит речь (от 0 до 1). Высокое значение говорит о том, что в треке много речи или вокала (например, в рэпе или подкастах). (число, float)

acousticness – Параметр, указывающий на "акустичность" трека (от 0 до 1). Высокое значение свидетельствует о том, что в треке преобладают акустические инструменты или натуральное звучание. (число, float)

instrumentalness – Параметр, показывающий, насколько трек является инструментальным (от 0 до 1). Чем ближе значение к 1, тем более в треке отсутствуют вокальные элементы. (число, float)

liveness – Параметр, указывающий на присутствие "живой" записи или аудитории в треке (от 0 до 1). Высокие значения означают, что трек был записан с живым выступлением или с аудиторией. (число, float)

valence – Параметр, измеряющий эмоциональную тональность трека (от 0 до 1). Значение близкое к 1 указывает на позитивные эмоции в треке, в то время как значение ближе к 0 — на негативные. (число, float)

tempo – Темп трека в ударах в минуту (BPM). Это число указывает, как быстро или медленно звучит трек. (число)

time_signature – Размер такта трека, обычно 4/4, 3/4 и так далее. Это число обозначает, сколько долей в такте. (число)

track_genre – Жанр трека, как классифицированный в базе данных Spotify (например, рок, поп, джаз и т.д.). (строка)

### Гипотеза 1:
Популярность трека зависит от "скрытых" связей, существующих между обьективными музыкальными параметрами. Скажем, определенная совокупность темпа, шумности, жанра гарантирует треку всенародную любовь и признание. Это может обосновывать отсутствие значимых корреляций

### Гипотеза 2:
Популярность трека является, скорее, залогом известного имени. Трек с большей вероятностью станет популярным, если он был выпущен супер-звездой, особенно - если внутри успешного музыкального альбома

## threshold популярности

Добавим параметр, отражающий то, переходит ли трек медианный параметр популярности. Он будет равен 1, если популярность выше 22 - и 0 если нет.

```python
soundframe['popularity_threshold'] = (soundframe['popularity'] > soundframe['popularity'].median()).astype(int)
print(soundframe['popularity_threshold'].head())
```

Посмотрим, в каких жанрах чаще всего треки переходят линию отсечки

```python
dataframes = {
    'Rock': r_sf_rock,
    'Dance': r_sf_dance,
    'World Music': r_sf_world_music,
    'Ambient': r_sf_ambient,
    'Comedy': r_sf_comedy,
    'Jazz': r_sf_jazz,
    'Emotional': r_sf_emotional,
    'Party': r_sf_party
}
percent_over_threshold = {}

for genre, df in dataframes.items():
    if 'popularity_threshold' in df.columns:
        percent_over_threshold[genre] = df['popularity_threshold'].mean() * 100
    else:
        print(f"Столбец 'popularity_threshold' не найден в {genre}")
        percent_over_threshold[genre] = None  # или можно пропустить жанр

# Убираем жанры с None значениями
percent_over_threshold = {k: v for k, v in percent_over_threshold.items() if v is not None}

plt.figure(figsize=(10, 6))
plt.bar(percent_over_threshold.keys(), percent_over_threshold.values(), color='skyblue')
plt.xlabel('Жанры')
plt.ylabel('% треков, пересекших порог')
plt.title('Процент треков, пересекших порог для разных жанров')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

#### Как выяснилось, не смотря на то, что рок оказался самым популярным жанром, наибольшее число "качественных" треков оказалось именно среди мировой музыки

```python
nn_model = Sequential()
nn_model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
nn_model.add(Dense(64, activation='relu'))
nn_model.add(Dense(1, activation='linear'))
nn_model.compile(optimizer='adam', loss='mean_squared_error')
```

Создана нейронная сеть с архитектурой, состоящей из трех полносвязных слоев. Первый слой содержит 128 нейронов с активацией ReLU, второй слой — 64 нейрона с той же активацией, а выходной слой имеет один нейрон с линейной активацией. Модель скомпилирована с использованием оптимизатора Adam и функции потерь MSE (среднеквадратичная ошибка).

```python
history = nn_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

Модель нейронной сети обучалась на обучающей выборке в течение 100 эпох с размером мини-пакета 32. В процессе обучения использовались данные валидации (тестовая выборка) для отслеживания производительности модели на независимых данных.

```python
loss = nn_model.evaluate(X_test, y_test)
print(f'Средняя квадратичная ошибка на тестовых данных: {loss}')
y_pred = nn_model.predict(X_test)
```

## КРОСС-ВАЛИДАЦИЯ

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
results = {}

lr = LinearRegression()
lr_scores = cross_val_score(lr, X, y, cv=kf, scoring='neg_mean_squared_error')
results['Linear Regression'] = -lr_scores

def create_nn_model():
    model = Sequential()
    model.add(Dense(64, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

nn_model = KerasRegressor(build_fn=create_nn_model, epochs=50, batch_size=32, verbose=0)
nn_scores = cross_val_score(nn_model, X, y, cv=kf, scoring='neg_mean_squared_error')
results['Neural Network'] = -nn_scores

tree = DecisionTreeRegressor(min_samples_split = 4, min_samples_leaf=7, random_state=42)
tree_scores = cross_val_score(tree, X, y, cv=kf, scoring='neg_mean_squared_error')
results['Decision Tree'] = -tree_scores

gbr = GradientBoostingRegressor(n_estimators = 300, learning_rate=0.2, max_depth=7, random_state=42)
gbr_scores = cross_val_score(gbr, X, y, cv=kf, scoring='neg_mean_squared_error')
results['Gradient Boosting'] = -gbr_scores
```

В данном блоке выполнена 5-кратная кросс-валидация для линейной регрессии, дерева решений и градиентного бустинга с оценкой по метрике MSE. Результаты сохранены в словаре results для дальнейшего анализа.

```python
plt.figure(figsize=(12, 8))

# Гистограмма для Линейной регрессии
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.hist(lr_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title('Linear Regression MSE')
plt.xlabel('MSE')
plt.ylabel('Frequency')

# Гистограмма для Нейронной сети
plt.subplot(2, 2, 2)
plt.hist(nn_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title('Neural Network MSE')
plt.xlabel('MSE')
plt.ylabel('Frequency')

# Гистограмма для Решающих деревьев
plt.subplot(2, 2, 3)
plt.hist(tree_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title('Decision Tree MSE')
plt.xlabel('MSE')
plt.ylabel('Frequency')

# Гистограмма для Градиентного бустинга
plt.subplot(2, 2, 4)
plt.hist(gbr_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title('Gradient Boosting MSE')
plt.xlabel('MSE')
plt.ylabel('Frequency')

# Подстроиваем макет
plt.tight_layout()
plt.show()
```

Linear Regression: Гистограмма показывает распределение MSE для линейной регрессии. Ошибки имеют небольшой разброс, что указывает на стабильность модели, но могут быть более высокими по сравнению с другими методами.

Decision Tree: Распределение ошибок для решающих деревьев значительно шире. Это свидетельствует о том, что модель может переобучаться на определенных частях данных, показывая менее стабильные результаты.

Gradient Boosting: Ошибки для градиентного бустинга демонстрируют суженное распределение с низкими значениями MSE. Это указывает на то, что модель хорошо справляется с задачей и показывает высокий уровень точности.
