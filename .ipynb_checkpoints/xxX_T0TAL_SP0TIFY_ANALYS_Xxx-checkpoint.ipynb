Датасет разбит на жанры, содержит такие значимые параметры как имя артиста, название альбома, параметр популярности, длительности, наличие нецензурной лексики, музыкальные параметры выраженные численно, количество долей и жанр песни.

track_id – Уникальный идентификатор трека в Spotify. Это строка, которая однозначно идентифицирует трек в базе данных Spotify.

artists – Исполнители трека(строка).

album_name – Название альбома, в котором был выпущен трек(строка).

track_name – Название самого трека(строка).

popularity – Популярность трека на Spotify, оцененная от 0 до 100. Чем выше значение, тем популярнее трек среди пользователей платформы(число).

duration_ms – Длительность трека в миллисекундах(число).

explicit – Флаг, указывающий на наличие нецензурной лексики в треке. Значение True означает, что трек содержит ненормативную лексику, а False — нет.(bool)

danceability – Параметр, описывающий, насколько трек удобен для танцев (от 0 до 1). Чем выше значение, тем легче танцевать под этот трек.(число, float)

energy – Энергичность трека (от 0 до 1). Чем выше значение, тем более динамичный и интенсивный трек.(число, float)

key – Ключ (тональность) трека, выраженный числом от 0 до 11. Это относится к музыкальной шкале и указывает на главную тональность трека (например, 0 – это C, 1 – C#/Db и т.д.).(число)

loudness – Общая громкость трека в децибелах (dB). Это значение отражает громкость, при которой трек был записан и воспроизводится.(отрицательное число, float)

mode – Модальность трека, которая определяет его тональность как мажорную (1) или минорную (0). (int bool)

speechiness – Параметр, указывающий, насколько трек содержит речь (от 0 до 1). Высокое значение говорит о том, что в треке много речи или вокала (например, в рэпе или подкастах). (число, float)

acousticness – Параметр, указывающий на "акустичность" трека (от 0 до 1). Высокое значение свидетельствует о том, что в треке преобладают акустические инструменты или натуральное звучание. (число, float)

instrumentalness – Параметр, показывающий, насколько трек является инструментальным (от 0 до 1). Чем ближе значение к 1, тем более в треке отсутствуют вокальные элементы. (число, float)

liveness – Параметр, указывающий на присутствие "живой" записи или аудитории в треке (от 0 до 1). Высокие значения означают, что трек был записан с живым выступлением или с аудиторией. (число, float)

valence – Параметр, измеряющий эмоциональную тональность трека (от 0 до 1). Значение близкое к 1 указывает на позитивные эмоции в треке, в то время как значение ближе к 0 — на негативные. (число, float)

tempo – Темп трека в ударах в минуту (BPM). Это число указывает, как быстро или медленно звучит трек. (число)

time_signature – Размер такта трека, обычно 4/4, 3/4 и так далее. Это число обозначает, сколько долей в такте. (число)

track_genre – Жанр трека, как классифицированный в базе данных Spotify (например, рок, поп, джаз и т.д.). (строка)

### threshold популярности

Добавим параметр, отражающий то, переходит ли трек медианный параметр популярности. Он будет равен 1, если популярность выше 22 - и 0 если нет.

```python
soundframe['popularity_threshold'] = (soundframe['popularity'] > soundframe['popularity'].median()).astype(int)
print(soundframe['popularity_threshold'].head())
```

#### Параметр основан на целевом признаке, что по идее не очень хорошо, однако будет очень любопытно оценить его влияние на конечный результат. Предполагается, что он не слишком сильно повлияет на результат задачи линейной регрессии

### Для улучшения восприятия добавим параметр длительности в секундах

```python
soundframe['duration_sec'] = soundframe['duration_ms'] / 10000
```

### Solo author или music band?

Можно сделать предположение, что имена исполнителей чаще записываются в виде двух слов(имени, отчества в англо-американской традиции), тогда как музыкальные группы - одно запоминающеся слово(если убрать артикль 'THE' )
Для начала добавим просто показатель, содержащий количество слов в названии. Беглый анализ датасета показывает, что зачастую в качестве разделителей в названии используется либо ' ', либо ';'

```python
soundframe['artist_len'] = soundframe['artists'].fillna('').apply(lambda x: len(x.replace(';', ' ').split()) if isinstance(x, str) else 0)
print(soundframe['artist_len'].head(10))
```

```python
grouped = soundframe.groupby('artist_len').count()
```

```python
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(grouped['artists'], labels=None, autopct='')

plt.legend(wedges, [f'{grouped.index[i]}: {grouped["artists"].iloc[i]} ({grouped["artists"].iloc[i] / grouped["artists"].sum() * 100:.1f}%)' for i in range(len(grouped))], title="Artist Count", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Распределение по количеству артистов')

plt.show()
```

#### В целом предположение подтверждается, явный перевес в пользу исполнителей с названием из двух слов свидетельствует о том, что вероятнее всего это их first и second name, можно сделать дискриминационный показатель, который будет равен 1, если artists name имеет 2 слова

```python
soundframe['solo_artist'] = (soundframe['artist_len'] == 2).astype(int)
print(soundframe['solo_artist'].head(10))
```

### Альбом - не альбом?

#### В блоке, посвященном OHE - преобразованию мы обсудим целесообразность преобразовывать к OHE-векторам размерности 10000+ названия альбомов и исполнителей. Пока что попробуем сгладить углы и соотнесём альбомы с числовым признаком, показывающим количество треков в каждом.

# ЭКСПЕРИМЕНТЫ С НЕЙРОННЫМИ СЕТЯМИ и машинным обучением*

## Преобразование к One Hot Encoding


### О OHE-преобразовании, оценка целесообразности

#### Нейронные сети, модели машинного обучения не умеют обрабатывать категориальные текстовые признаки, имена и фразы, слова и предложения "из коробки". Для работы с ними существует 2 самых популярных метода - плотные вектора и One Hot Encoding. Так же коммандой были рассмотрены методы Ordinal Encoder и Target Encoded

### Плотные вектора, плотное векторное представление

#### Метод, при котором слова, буквенные последовательности у нас представляются координатами в N-мерном пространстве. Очень красивый и элегантный способ, нейронная сеть исходя из собственных задач сама размещает слова на своём месте. В языковых пространствах больших языковых моделей слова King и Queen имеют одинаковый оффсет от слов Man и Woman соответственно. И этот прекрасный, элегантный способ нам совершенно не подходит!

Обычно языковые пространства намеренно ограничивают самыми популярными словами. Рядовой вокабуляр человека относительно ограничен. В задаче, где нам надо классифицировать отзывы на хорошие и плохие даже 2-3 тысяч слов будет вполне достаточно, но что делать с названиями групп, альбомов? Каждое из них уникально, зачастую не несёт никакой смысловой ценности.
С использованием плотных векторов задача будет решаема только с использованием обучения на тестовых данных, и такому решению грош цена.

#### Отклонён по причине: необходимости обучения на тестовых данных.

### ONE HOT ENCODING, кодирование с использованием единичных векторов векторов размерности, равной размерности признакового пространства.


#### Метод, при котором уникальные слова, цепочки слов кодируются при помощи уникальных единичных векторов формата (0, 1, 0). С этим форматом могут работать очень многие модели машинного обучения, разумеется не без оговорок.

При использовании OHE-кодирования вылезает одна главная неприятная проблема. У нас - 30 тысяч уникальных исполнителей. Значит вектор, кодирующий каждого исполнителя будет иметь размерность 30 тысяч. Это является отвратительно неоптимальным способом использовать компьютерные ресурсы, однако, похоже, у нас нет других альтернатив.

Ещё одна проблема OHE - кодирования - большинство кодировщиков используют классические нампай векторы, которые нельзя просто взять и положить внутрь датафрейма. Их придется скармливать в нейросеть в каком-либо другом формате. Ещё вариант - использовать встроенный пандасовский get_dummies. Он создаст нам 30 тысяч колонок с исполнителями, 10 тысяч колонок с альбомами и около 100 колонок с жанрами. 

#### Компромисно, с оговорками - принимаем. Лучше, чем ничего. Кодировать будем признак с минимальной размерностью - жанр.

### Target Encoded

#### Метод кодирования категориальных переменных, при котором каждой категории присваивается значение на основе статистики целевой переменной.
Был отвергнут коммандой по причине "читерство", делать так как будто не есть хорошо

### Ordinal encoder, кодирование с использованием единичных векторов векторов размерности, равной размерности признакового пространства.


#### Метод преобразования категориальных переменных с естественным порядком или ранжированием в числовые значения. Он используется в тех случаях, когда категории имеют внутреннюю упорядоченность, например: "Низкий", "Средний", "Высокий"

У метода есть 2 явных преимущества по сравнению с OHE
Эффективность: Использует меньше памяти по сравнению с one-hot encoding. (ВЕРНО ДЛЯ МАЛЫХ ВЫБОРОК)
Простота: Легко применимо, когда категориальные данные имеют явный порядок.

К сожалению, у нас тут не рейтинг песни от 1 до 5, где метод подошёл бы прекрасно. У нас - 100 уникальных жанров.
#### Увы - не наш случай.

### Преобразуем категориальный признак жанра через get_dummies, оставим только числовые значения

#### Не смотря на все противоречия, команда остановилась именно на нём. Метод - кривой, косой, неоптимальный - но безальтернативный. Мы не можем адекватно всунуть в датафрейм нампай-вектор, а остальные решения оказались не столь подходящими.

```python
soundframe = pd.get_dummies(soundframe, columns = ['track_genre'])
for col in soundframe.columns:
    if 'genre_' in col:
        soundframe[col] = soundframe[col].astype(int)
numeric_soundframe = soundframe.select_dtypes(include=[int, float])
```

Пока что сохраним параметр popularity_threshold, посмотрим на разницу в задаче линейной регрессии

Разумно было бы поделить данные исходя из их жанра, т.к. совокупность обьективных музыкальных параметров зачастую неразрывно с этим связана, но пока что попробуем так

Так же будем использовать standart scaler из за высокого разброса числовых признаков

### На графике существует значимое плато, значит в модели разумно применить регуляризацию!

#### Пройдемся по основным методам регуляризации.

#### PCA - метод главных компонент, используется для оптимизации(снижения размерности), с ним результат станет только хуже, хоть и заметно быстрее - ибо обрезать у модели мы можем достаточно.

#### R1 регуляризация, представленная LASSO - регрессией ->  В этой модели регуляризация добавляет штраф, пропорциональный абсолютной величине коэффициентов. Это может привести к обнулению некоторых коэффициентов, что эффективно отбирает признаки и приводит к более простым моделям.

#### R2 регуляризация, представленная RIDGE - регрессией -> В этом методе регуляризации добавляется штраф пропорциональный квадрату величины коэффициентов. Это ограничивает значения коэффициентов, чтобы они не становились слишком большими.

```python
ridge = Ridge(alpha=1.0)  # Альфа - гипер-параметр лямбда.
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
r2_ridge = r2_score(y_test, y_pred_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)

print("Ridge Regression R2 Score:", r2_ridge)
print("Root Mean Squared Error (Ridge):", rmse_ridge)
print("Mean Absolute Error (Ridge):", mae_ridge)
```

```python
lasso = Lasso()
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)

rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))
r2_lasso = r2_score(y_test, y_pred_lasso)
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)

print("Lasso Regression R2 Score:", r2_lasso)
print("Root Mean Squared Error (Lasso):", rmse_lasso)
print("Mean Absolute Error (Lasso):", mae_lasso)
```

#### Кросс - валидация для указанных моделей

```python
ridge = Ridge()
lasso = Lasso()

cv = KFold(n_splits=5, shuffle=True, random_state=42)

ridge_r2_scores = []
ridge_rmse_scores = []
ridge_mae_scores = []

lasso_r2_scores = []
lasso_rmse_scores = []
lasso_mae_scores = []

for fold, (train_index, test_index) in enumerate(cv.split(X_train), 1):
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

    ridge.fit(X_train_fold, y_train_fold)
    y_pred_ridge = ridge.predict(X_test_fold)

    lasso.fit(X_train_fold, y_train_fold)
    y_pred_lasso = lasso.predict(X_test_fold)

    ridge_r2 = r2_score(y_test_fold, y_pred_ridge)
    ridge_rmse = np.sqrt(mean_squared_error(y_test_fold, y_pred_ridge))
    ridge_mae = mean_absolute_error(y_test_fold, y_pred_ridge)

    lasso_r2 = r2_score(y_test_fold, y_pred_lasso)
    lasso_rmse = np.sqrt(mean_squared_error(y_test_fold, y_pred_lasso))
    lasso_mae = mean_absolute_error(y_test_fold, y_pred_lasso)

    ridge_r2_scores.append(ridge_r2)
    ridge_rmse_scores.append(ridge_rmse)
    ridge_mae_scores.append(ridge_mae)

    lasso_r2_scores.append(lasso_r2)
    lasso_rmse_scores.append(lasso_rmse)
    lasso_mae_scores.append(lasso_mae)

    print(f"Fold {fold} - Ridge:")
    print(f"  R2: {ridge_r2:.4f}")
    print(f"  RMSE: {ridge_rmse:.4f}")
    print(f"  MAE: {ridge_mae:.4f}")
    print("-" * 40)

    print(f"Fold {fold} - Lasso:")
    print(f"  R2: {lasso_r2:.4f}")
    print(f"  RMSE: {lasso_rmse:.4f}")
    print(f"  MAE: {lasso_mae:.4f}")
    print("-" * 40)

print("\nAverage results for Ridge:")
print(f"Average R2: {np.mean(ridge_r2_scores):.4f}")
print(f"Average RMSE: {np.mean(ridge_rmse_scores):.4f}")
print(f"Average MAE: {np.mean(ridge_mae_scores):.4f}")

print("\nAverage results for Lasso:")
print(f"Average R2: {np.mean(lasso_r2_scores):.4f}")
print(f"Average RMSE: {np.mean(lasso_rmse_scores):.4f}")
print(f"Average MAE: {np.mean(lasso_mae_scores):.4f}")
```

#### Мы произвели экспресс-оценку эффективности модели, давайте выполним полную оценку модели с подобранными гипер-параметрами

```python
gbr = GradientBoostingRegressor(n_estimators=300, learning_rate=0.2, max_depth=7, random_state=42)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

mae_cv = []
r2_cv = []
rmse_cv = []

for fold, (train_index, test_index) in enumerate(kf.split(X_train), 1):
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

    gbr.fit(X_train_fold, y_train_fold)

    y_pred_fold = gbr.predict(X_test_fold)

    mae = mean_absolute_error(y_test_fold, y_pred_fold)
    r2 = r2_score(y_test_fold, y_pred_fold)
    rmse = np.sqrt(mean_squared_error(y_test_fold, y_pred_fold))

    mae_cv.append(mae)
    r2_cv.append(r2)
    rmse_cv.append(rmse)

    print(f"Fold {fold}:")
    print(f"  MAE: {mae:.4f}")
    print(f"  R2: {r2:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print("-" * 40)

print("\nCross-Validation Results (Averaged):")
print(f"Cross-Validation MAE: {np.mean(mae_cv):.4f}")
print(f"Cross-Validation R2: {np.mean(r2_cv):.4f}")
print(f"Cross-Validation RMSE: {np.mean(rmse_cv):.4f}")

feature_variance = np.var(X_train, axis=0)
print("\nFeature Variance:", feature_variance)
```

#### КРОСС-ВАЛИДАЦИЯ НЕЙРОННОЙ СЕТИ

#### Выполним кросс-валидацию нашей простенькой модели, не возлагая больших результатов взглянем на полученные метрики.

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

def create_nn_model():
    model = Sequential()
    model.add(Dense(64, input_dim=X_train_scaled.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='huber', metrics=['mean_absolute_error', 'mean_squared_error'])
    return model

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

model = create_nn_model()
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, verbose=0)

y_pred = model.predict(X_test_scaled)

nn_mae = mean_absolute_error(y_test, y_pred)
rmse_value = rmse(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nNeural Network")
print(f"MAE на валидационной выборке: {nn_mae:.4f}")
print(f"RMSE на валидационной выборке: {rmse_value:.4f}")
print(f"R2 на валидационной выборке: {r2:.4f}")
```
